```python
import xml.etree.ElementTree as ET
tree = ET.parse('../Mockup1/mockup1.ui')
root = tree.getroot()
for elem in tree.iter():
    print("{}: {} => {}".format(elem.tag,elem.attrib,elem.text))
```

```python
def widget(elem):
    if(elem.tag == 'widget'):
        name = elem.attrib['name']
        qclass = elem.attrib['class']
        return "{} {} = new {}(this);\n".format(qclass,name,qclass)


output = ""

for elem in tree.iter():
    output += widget(elem)

print(output)


```
