```python
import xml.etree.ElementTree as ET
tree = ET.parse('calculatorform.ui')
root = tree.getroot()
```

```python
output = ""
for elem in tree.iter():
    if(elem.tag == 'widget'):
        name = elem.attrib['name']
        qclass = elem.attrib['class']
        output.append("{} {} = new {}(this);\n".format(qclass,name,qclass))
    if(elem.tag == 'x'):
        x = elem.text
    if(elem.tag == 'y')
        y = elem.text
        output.append("{}->move({},{});\n".format(name,x,y))
print(output)
```
